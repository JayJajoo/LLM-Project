```python
from dotenv import load_dotenv
from langchain_community.vectorstores import Chroma
from langchain_openai.embeddings import OpenAIEmbeddings
import warnings
import os
warnings.filterwarnings("ignore")
load_dotenv()
```

```python
vector_db_path = "D:/LLM projects/LLM Project/backend/agent/VectorDB"
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")

regular_courses_database = Chroma(collection_name="RegularCourses",embedding_function=embeddings,persist_directory = vector_db_path)
special_topics_database = Chroma(collection_name="SpecialCourses",embedding_function=embeddings,persist_directory= vector_db_path)
```

```python
reg_course_data = regular_courses_database.get()
special_course_data = special_topics_database.get()
```

```python
reg_course_data["metadatas"][0]
```

```python
colleges = [course['college'] for course in reg_course_data["metadatas"]] + [course['college'] for course in special_course_data["metadatas"]]
colleges = set(colleges)
colleges  = [{"value":value} for value in colleges]
colleges
```

```python
departments = [course['department'] for course in reg_course_data["metadatas"]] + [course['department'] for course in special_course_data["metadatas"]]
departments = set(departments)
departments = [{"value":value} for value in departments]
departments
```

```python
mp = {}
courses = []

for course in reg_course_data["metadatas"]:
    if mp.get(course["course_number"])!=None:
        continue
    else:
        mp[course["course_number"]] = 1
        courses.append({"value":course["course_number"] , "misc_data":course["title"]})

for course in special_course_data["metadatas"]:
    if mp.get(course["course_number"])!=None:
        continue
    else:
        mp[course["course_number"]] = 1
        courses.append({"value":course["course_number"] , "misc_data":course["title"]})
```

```python
courses
```

```python
import json


for key,val in {"courses":courses,"departments":departments,"colleges":colleges}.items():
    with open(f'{key}.json', 'w') as f:
        json.dump(val, f)
```
